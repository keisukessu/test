# 関数
* 再利用可能なコードの塊を**関数** (function)と呼ぶ
* 関数はプログラム中から呼び出されなければ実行されない
* 関数の特徴：
 * 名前（関数名）を持つ
 * 引数を持つ（0以上）
 * ドキュメンテーション文字列を持つ（なくてもいいけど...）
 * 本文を持つ
 * なにかを返す（return）




図1.svg

````python
def is_even(i):
  """
  入力: 正の整数i
  出力: iが偶数ならTrueをそうでなければFalseを返す
  """
  return i%2==0

is_even(5)
````

# return文の意味

* returnがなかったらその関数は`None`を返す

````python
def is_even(i):
  """
  入力: 正の整数i
  出力: iが偶数ならTrueをそうでなければFalseを返す？？
  """
  i%2==0

print(is_even(5))
````

````python
def func_a():
  print('inside func_a')
def func_b(y):
  print('inside func_b')
  return y
def func_c(z):
  print('inside func_c')
  return z()

print(func_a())
print(5+func_b(2))
print(func_c(func_a))
````

# 関数あれこれ

## 引数はいくつでも…
```
def print_name(first, last, reverse):
  if reverse:
    print(last, first)
  else:
    print(first, last)
```

## デフォルト引数というのが作れる…
```
def print_name(first, last, reverse=True):
  if reverse:
    print(last, first)
  else:
    print(first, last)
```

## キーワード引数というのを使うと特定の引数を指定できる…
```
def print_name(first, last, reverse=True):
  if reverse:
    print(last, first)
  else:
    print(first, last)

print_name("Kohei", "Matsumura")
print_name("Matsumura", "Kohei")
print_name(first="Kohei", last="Matsumura")
print_name(last="Matsumura", first="Kohei")
```


````python
def print_name(first, last, reverse):
  if reverse:
    print(last, first)
  else:
    print(first, last)

print_name("Kohei","Matsumura")
````
