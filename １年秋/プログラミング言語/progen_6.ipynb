# 辞書におけるkeyとvalue

* key（キー）
 * ユニークでなければならない（同じkeyは2つ存在できない）
 * 不変型でなければならない（int, float, string, tuple, bool）

* value（値）
 * どんな型でもよい（不変型、可変型）
 * 重複していても良い
 * 可変型がOKなので、リストや辞書を値として持っても良い

# リストと辞書

* リスト
  * 順序を伴った要素の集合
  * 整数を用いた添字により要素にアクセスできる
  * **順序**が保持される
  * 整数のみが添え字に使える

* 辞書
  * **key**と**value**の結びつき
  * keyを用いた添字により要素にアクセスできる
  * **順序**は保証されない
  * 添字（key）にはあらゆる不変型を使える

# 辞書を用いたフィボナッチ数列

```
def fib_with_dict(x, d):
  if x in d:
    return d[x]
  else:
    ans = fib_with_dict(n-1, d) + fib_with_dict(n-2, d)
    d[n] = ans
    return ans

d = {0:1, 1:1}
print(fib_with_dict(8, d))
```

* まず、辞書を調べて結果がすでにないかをチェック
 * あれば計算せずに結果を返す
 * なければ計算し、その結果を辞書に追加する
* この例のように計算結果を記録しておいて再利用する方法を**メモ化**と言う

````python

def fib(x):
    global cnt
    cnt+=1
    if x==0 or x==1:
        return 1
    else:
        return fib(x-1) + fib(x-2)
cnt = 0
print(fib(34))
print(cnt)
````

````python
def fib_with_dict(x, d):
  global cnt
  cnt+=1
  if x in d:
    return d[x]
  else:
    ans = fib_with_dict(x-1, d) + fib_with_dict(x-2, d)
    d[x] = ans
    return ans

cnt = 0
d = {0:1, 1:1}
print(fib_with_dict(33, d))
print(cnt)
````

# どれくらい効率化されるか

* fib(34)を呼び出した場合、再帰呼び出しの数は18,454,929回
* fib_with_dict(34)を呼び出した場合再帰呼び出しの数は67回
* `x=34`の例だと30万倍くらい効率的
 * 辞書を用いたメモ化は非常に効率的になる場合がある
* ただし、メモ化などによる副作用（この部分にかかるオーバーヘッド）を考慮する必要もある

# 楽曲の歌詞を分析する

1. `{str:int}`の辞書を使って、歌詞に使われている単語の頻度を記録する
2. 最も使われている単語とその頻度を探す
  * 単語が複数あるかもしれないのでリストを使う
  * `(単語リスト, 頻度)`というタプルを返す　`(list, int)`
3. 最低x回使われている単語を探す
  * 引数を使って頻度を指定可能にする
  * 頻度に応じて単語のリストを返す `[(list, int), (list, int)... ]`

````python
yesterday = """
Yesterday
All my troubles seemed so far away
Now it looks as though they’re here to stay
Oh, I believe in yesterday
Suddenly
I’m not half the man I used to be
There’s a shadow hanging over me
Oh, yesterday came suddenly
Why she had to go
I don’t know, she wouldn’t say
I said something wrong
Now I long for yesterday
Yesterday
Love was such an easy game to play
Now I need a place to hide away
Oh, I believe in yesterday
Why she had to go
I don’t know, she wouldn’t say
I said something wrong
Now I long for yesterday
Yesterday
Love was such an easy game to play
Now I need a place to hide away
Oh, I believe in yesterday
""".lower().split()
````

````python
def lyrics_to_frequencies(lyrics):
    myDict = {}
    for word in lyrics:
        if word in myDict:
            myDict[word] += 1
        else:
            myDict[word] = 1
    return myDict
````

````python
beatles = lyrics_to_frequencies(yesterday)
beatles
````

````python
def most_common_words(freqs):
    values = freqs.values()
    best = max(freqs.values())
    words = []
    for k,v in freqs.items():
        if v == best:
            words.append(k)
    return (words, best)
````

````python
most_common_words(beatles)
````

````python
def words_often(freqs, minTimes):
    result = []
    done = False
    while not done:
        temp = most_common_words(freqs)
        if temp[1] >= minTimes:
            result.append(temp)
            for w in temp[0]:
                del(freqs[w])
        else:
            done = True
    return result
````

````python
print(words_often(beatles, 4))
````
