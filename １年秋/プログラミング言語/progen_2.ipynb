# 文字の参照
`[<数字>]`を用いることで文字列から<数字>で示される部分の文字を取り出すことができる

* `s = "abc"`であれば数字は-3からから2の間で指定できる。
* 負ではない整数は0から2まで（始端が0）
* 負の整数は-1から-3まで（終端が-1）

|  s[0] |  s[1]  | s[2] | s[-1]| s[-2] | s[-3]|
| ---- | ---- | ---- | ---- | ---- | ---- |
|  a  |  b  | c | c | b | a |



#文字列のスライス

* `[<開始>:<終了>:<ステップ>]`を用いることで文字列の一部を切り出した部分文字列を作ることができる
* `[<開始>:<終了>]`を指定した場合はステップは1となる
* <開始>や<終了>の数字を指定しないこともできる

## `s = "abcdefgh"`での例

* `s[3:6]`（`s[3:6:1]`に等しい）→ `"def"`
* `s[3:6:2] ` → `"df"`
* `s[::]` → "abcdefgh" （`s[0:len(s):1]`も同じ）
* `s[::-1]` → `"hgfedbca"` （`s[-1:-(len(s)+1):-1]`も同じ））
* `s[4:1:-2]` → `"ec"`

````python
s = "abcdefgh"
s[3:6]
s[3:6:2]
s[::]
s[::-1]
s[4:1:-2]
````

# 試行錯誤法（総当たり法））

* 何らかの問題があるとき
* 正解と思われる候補を挙げる
* 候補が正解であるかを確かめる
* 正解に辿り着くか、候補がなくなったら止まる

````python
cube = 8
for guess in range(cube+1):
  if guess**3 == cube:
    print(cube, "の立法根は", guess)
````

````python
#完全立法の立法根を求める
x = int(input('Enter an integer: '))
ans = 0
while ans**3 < abs(x):
  ans = ans + 1
if ans**3 != abs(x):
  print(x, "is not a perfect cube")
else:
  if x<0:
    ans = -ans
  print("Cube root of", x, "is", ans) 
````

# 近似解

* 実用上十分な解を求める方法

* （完全立法でない）`cube`の立法根を求める
  * ある候補`guess`からスタートして（とても）小さな数`increment`を足していく
  * `|guess**3 - cube| >= epsilon`の間↑を繰り返す

* `increment`を小さくすれば解の精度は高まるがプログラムは遅くなる
* `epsilon`を大きくすればプログラムは早くなるが解の精度は下がる


````python
cube = 27
epsilon = 0.001
guess = 0.0
increment = 0.001
num_guesses = 0
while abs(guess**3 - cube) >= epsilon and guess <= cube:
  guess += increment
  num_guesses += 1
print('num_guesses =', num_guesses)
if abs(guess**3 - cube) >= epsilon:
    print('Failed on cube root of', cube)
else:
  print(guess, 'is close to the cube root of', cube)
````

# 2分法

* 一回の繰り返しのうちで探索空間を1/2にしていく
* `guess`を中間におく
* 繰り返す

ちょっと遊んでみようか？？

````python
cube = 27
epsilon = 0.001
low = 0
high = cube
guess = (low+high)/2.0
num_guesses = 0
while abs(guess**3 - cube) >= epsilon:
  if guess**3 < cube:
    low = guess
  else:
    high = guess
  guess = (low+high)/2.0
  num_guesses += 1
print('num_guesses =', num_guesses)
print(guess, 'is close to the cube root of', cube)
````

# 2分法

* 総当たり法に比べて探索空間が小さくできる

* 探索空間
  * 1回目:N/2
  * 2回目:N/4
  * k回目:N/2^k
